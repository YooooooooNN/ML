```python
app_train.describe()
```

```python
int_features = []
float_features = []
object_features = []
for dtype, feature in zip(app_train.dtypes, app_train.columns):
    if dtype == 'float64':
        float_features.append(feature)
    elif dtype == 'int64':
        int_features.append(feature)
    else:
        object_features.append(feature)
```

```python
app_train[int_features].describe()
app_train[int_features].mean()
app_train[int_features].std()
```

```python
app_train[float_features].describe()
# app_train[float_features].mean()
# app_train[float_features].std()
```

```python
app_train[object_features].describe()
# app_train[object_features].mean()
# app_train[object_features].std()
```

# <font color = blue>計算某一欄的的mean/std

```python
print('REGION_POPULATION_RELATIVE mean value:',app_train['REGION_POPULATION_RELATIVE'].mean())
print('REGION_POPULATION_RELATIVE std value:',app_train['REGION_POPULATION_RELATIVE'].std())
```

# 畫直方圖、長條圖

```python
# app_train['OWN_CAR_AGE'].plot.hist(alpha = 0.6)
app_train['OWN_CAR_AGE'].plot.hist(alpha = 0.6);
```

```python
plt.hist(app_train['OWN_CAR_AGE'],color = 'yellow',alpha = 0.5)
plt.title('OWN_CAR_AGE')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
plt.hist(app_train['TARGET']);
```

```python
plt.bar(['0','1'], app_train['TARGET'].value_counts(),color = 'green',alpha = 0.4);
#value_counts來計算各唯一值
```

```python
plt.hist(app_train['CODE_GENDER']);
plt.bar(['F','M','XNA'], app_train['CODE_GENDER'].value_counts(), color = 'gray')
plt.title('CODE GENDER')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```
